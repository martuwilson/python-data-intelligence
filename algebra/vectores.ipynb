```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
v1 = np.array([3, 4])
v2 = np.array([1, 2])

suma = v1 + v2
resta = v1 - v2
producto_escalar = np.dot(v1, v2)

print("Suma:", suma)
print("Resta:", resta)
print("Producto Escalar:", producto_escalar)
```

```python
fig, ax = plt.subplots()

ax.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='r', label='v1')
ax.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='b', label='v2')
ax.quiver(0, 0, suma[0], suma[1], angles='xy', scale_units='xy', scale=1, color='g', label='Suma')
ax.quiver(0, 0, resta[0], resta[1], angles='xy', scale_units='xy', scale=1, color='m', label='Resta')

ax.set_xlim(-1, 8)
ax.set_ylim(-1, 8)
ax.axhline(0, color='black',linewidth=0.5, ls='--')
ax.axvline(0, color='black',linewidth=0.5, ls='--')
ax.grid(True,which='both')

plt.legend()
plt.title('Visualización de Vectores')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
v1_3d = np.array([3, 4, 5])
v2_3d = np.array([1, 2, 3])

suma_3d = v1_3d + v2_3d
resta_3d = v1_3d - v2_3d
producto_escalar_3d = np.dot(v1_3d, v2_3d)

print("Suma 3D:", suma_3d)
print("Resta 3D:", resta_3d)
print("Producto Escalar 3D:", producto_escalar_3d)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.quiver(0, 0, 0, v1_3d[0], v1_3d[1], v1_3d[2], color='r', label='v1_3d')
ax.quiver(0, 0, 0, v2_3d[0], v2_3d[1], v2_3d[2], color='b', label='v2_3d')
ax.quiver(0, 0, 0, suma_3d[0], suma_3d[1], suma_3d[2], color='g', label='Suma 3D')
ax.quiver(0, 0, 0, resta_3d[0], resta_3d[1], resta_3d[2], color='m', label='Resta 3D')
ax.set_xlim([0, 8])
ax.set_ylim([0, 8])
ax.set_zlim([0, 8])
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
plt.title('Visualización de Vectores en 3D')
plt.legend()
plt.show()
```
