```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.DataFrame({
    "etiquetas": ['Manzanas','Bananas', 'Naranjas', 'Uvas'],
    "cantidad": [30,60,80,20]
})
df
```

```python
plt.pie(df['cantidad'])
plt.show()
```

```python
plt.pie(df['cantidad'], labels=df['etiquetas'])
plt.show()
```

```python
plt.pie(df['cantidad'], labels=df['etiquetas'], autopct='%1.1f%%')
plt.show()
```

```python
plt.pie(df['cantidad'],
        labels=df['etiquetas'],
        autopct='%1.1f%%',
       shadow=True,
       explode=(0,0,0,0.5))
plt.show()
```
