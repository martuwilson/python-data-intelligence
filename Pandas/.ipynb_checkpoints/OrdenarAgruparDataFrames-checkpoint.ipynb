```python
## Ordenación de DataFrames
df_ordenado = dt.sort_values(by="rating")
df_ordenado.head()
```

```python
df_ordenado = dt.sort_values(by="rating", ascending = False)
df_ordenado.head()
```

```python
## Psarle lista de las columnas que queremos
df_ordenado = dt.sort_values(by=["rating", "recaudación(M)"], ascending = False)
df_ordenado.head(10)
```

```python
df_agrupado = dt.groupby('género')['rating'].mean()
df_agrupado
```

```python
##Total recaudado x año
df_agrupado = dt.groupby('año')['recaudación(M)'].sum() ## agrupar todas las peliculas x año y una ves listo, hago suma de la recaudacion de cada grupo
df_agrupado.sort_values(ascending=False).head(10)
```
