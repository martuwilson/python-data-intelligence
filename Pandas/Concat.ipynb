```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
    'Nombre': ["Juan", "Gabriela", "Elena"], 
    'Edad': [23,31,21]
})
df1
```

```python
df2 = pd.DataFrame({
    'Nombre': ["Carmela", "Max", "Laura"], 
    'Edad': [34,25,29]
})
df2
```

```python
df_concat = pd.concat([df1,df2])
df_concat
```

```python
##por defector agrega verticalmente.

## horizontalmente es asi:
df_concat1 = pd.concat([df1,df2], axis=1)
df_concat1
```

```python
df_concat = pd.concat([df1,df2], ignore_index=True) ## en el de arriba repite los index numericos
df_concat
```

```python
##Agregar marcas para determinar que viene de cada DF
df_concat = pd.concat([df1,df2], keys=['df1','df2'])
df_concat
```

```python
##Agregar marcas para determinar que viene de cada DF
df_concat = pd.concat([df1,df2], keys=['Enero','Febrero'])
df_concat
```
