```python
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
df.head()
```

```python
## agrupar en hombres y mujeres
pasajeros = df.groupby('Sex').agg({
    'PassengerId':'count',
})

pasajeros
```

```python
## grafico con pandas
pasajeros.plot.bar(
    title='Cantidad de pasajeros por sexo',
    ylabel='Cantidad de pasajeros',
    xlabel='Sexo',
    width=0.4,
) ## pandas por detras usa matplotlib
```

```python
## grafico con Seaborn
import seaborn as sns
```

```python
pasajeros_sin_index = pasajeros.reset_index()
pasajeros_sin_index.head()
```

```python
sns.barplot(
    pasajeros_sin_index,
    y='PassengerId',
    x= 'Sex'
)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(
    pasajeros_sin_index['Sex'],
    pasajeros_sin_index['PassengerId']
)
```
