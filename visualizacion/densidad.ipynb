```python
import seaborn as sns
```

```python
df = pd.read_csv('./titanic.csv')
df = df.dropna(subset=['Age'], axis='index') # con index elimina la fila completa que tenga NaN
df
```

```python
sns.histplot(
    df,
    x='Age',
    kde=True,
    bins=15,
    stat='probability',
    hue='Sex'
)
```
