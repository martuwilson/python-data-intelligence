```python
import pandas as pd
```

```python
df = pd.read_csv('./tiendas_procesado.csv')
df
```

```python
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = (10, 5)
```

```python
plt.style.use(
    'bmh'
)

ax = df.plot.line()
##color de lineas
ax.xaxis.grid(
    color= 'blue',
    alpha=0.3, #opacidad
    linewidth=1, # grosor de linea
)

ax.yaxis.grid(
    color= 'red',
    alpha=0.3,
    linewidth=1,
    linestyle='--' # estilo de linea
)
```
