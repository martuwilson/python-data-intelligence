## Pie charts & Donuts

```python
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
df.head()
```

```python
pasajeros = df.groupby('Sex').agg({'PassengerId':'count'})
pasajeros
```

```python
pasajeros_sin_index = pasajeros.reset_index()
pasajeros_sin_index
```

```python
pasajeros.plot.pie(
    y='PassengerId',
    figsize=(5, 5),
    autopct='%1.1f%%',
    startangle=90, # comienza en 90 grados
    legend=False,
    title='Distribución de pasajeros por género',
    ylabel='')
```

```python
## Graficos con plotly
import plotly.express as px
```

```python
px.pie(
    pasajeros_sin_index,
    names='Sex',
    values='PassengerId',
    title='Distribución de pasajeros por género'
)
```

```python
## dona
fig = px.pie(
    pasajeros_sin_index,
    names='Sex',
    values='PassengerId',
    title='Distribución de pasajeros por género',
    hole=0.4
)

fig.show()
```
