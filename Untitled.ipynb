# Transformaci√≥n de los datos

```python
#creemos la columna ganancias
import random

def calcularGanancias(retweets):
    ganancia = retweets * random.randint(3,5)
    return ganancia

df["ganancias"] = df["retweets"].apply(calcularGanancias)

df.head()
```

```python
import random

def popularidad(fila):
    resultados = fila["followees"] / fila["followers"]
    return resultados

df["popularidad"] = df.apply(popularidad,axis=1) #con axis en 1 le especifico al apply que debe ser aplicado en cada fila del dataframe
df.head()
```
